```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
sns.set()
```

# Summary of the Problem

We are tasked to predict traffic on any given time (interval in hours) given the following information:

    1. Holiday
    2. Temperature
    3. Rain occured in the hour
    4. Snow occured in the hour
    5. Percentage of cloud cover
    6. Current weather
    7. Weather description
    8. Time of the data collected
    
 ## Initial thoughts
 
 What do we expect to see from this EDA?
 
     1. Current weather should be the most important feature
     2. Mean of each month's traffic volume should not have high variance since the number of cars own should be roughly the same
     3. Assuming normal ecnomic growth, traffic volume should have a Null Hypothesis for the ADF test, since economic growth should mean more cars being bought
     4. Traffic volume should spike during peak periods 

# Handling Data

```python
df = pd.read_csv('https://aisgaiap.blob.core.windows.net/aiap5-assessment-data/traffic_data.csv')
```

```python
df.head()
```

```python
# Check for missing data
for i in range(len(df.columns)):
    print(sum(df[df.columns[i]].isnull()))
```

```python
# Check Holidays
print(df['holiday'].value_counts())    
```

```python
# Breakup Date and Time into Day, DayofWeek, Month and Hour

df.date_time = pd.to_datetime(df.date_time)
```

```python
hour = df.date_time.dt.hour
dayofweek = df.date_time.dt.dayofweek
day = df.date_time.dt.day
month = df.date_time.dt.month

df['Month'] = month.copy() 
df['Hour'] = hour.copy()
df['DayOfWeek'] = dayofweek.copy()
df['Day'] = day.copy()
```

# Performing EDA

```python
# Plotting histogram for temperature

sns.distplot(df.temp)
temp_mean = np.mean(df.temp)
temp_var = np.var(df.temp)

print('Temperature Mean: ', temp_mean)
print('Temperature Variance: ', temp_var)
```

```python
# Plotting histogram for percentage of cloud cover 

sns.distplot(df.clouds_all)
cloudpcnt_mean = np.mean(df.clouds_all)
cloudpcnt_var = np.var(df.clouds_all)

print('Cloud Cover Percentage Mean: ', cloudpcnt_mean)
print('Cloud Cover Percentage Variance: ', cloudpcnt_var)
```

```python
# Plotting histogram for traffic volume

sns.distplot(df.traffic_volume)
traffic_mean = np.mean(df.traffic_volume)
traffic_var = np.var(df.traffic_volume)

print('Traffic Volume Mean: ', cloudpcnt_mean)
print('Traffic Colume Variance: ', cloudpcnt_var)
```

```python
# Plotting traffic volume as a time series

sns.tsplot(df.traffic_volume)
```

# Analysing data by date and time

```python
# Plotting data by mean of hour

hour_mean = df[['Hour','traffic_volume']].groupby(['Hour']).mean()
plt.plot(hour_mean)
plt.show()
plt.close()
```

```python
# Plotting data by mean of weekday

dayofweek_mean = df[['DayOfWeek','traffic_volume']].groupby(['DayOfWeek']).mean()
plt.plot(dayofweek_mean)
plt.show()
plt.close()
```

```python
# Plotting data by mean of weather main

weather_mean = df[['weather_main','traffic_volume']].groupby(['weather_main']).mean()
plt.figure(figsize=(10,8))
plt.plot(weather_mean)
plt.show()
plt.close()
```

```python
# Plotting data by mean of weather description

weather_desc_mean = df[['weather_main','traffic_volume']].groupby(['weather_main']).mean()
plt.figure(figsize=(10,8))
plt.plot(weather_desc_mean)
plt.show()
plt.close()
```

## Assumption: The mean of traffic volume would be roughly the same for every month
\
Regardless of hour and day of week, the amount of car usage should be roughly the same every month with minimal variance

```python
# Sum of traffic volume by month

plt.figure(figsize=(10,8))
plt.plot(df[['Month','traffic_volume']].groupby(['Month']).sum())
```

It seems like the variance is huge. The traffic volume surge at May and is the lowest in September.

## Possible Explanations
1. The seasonal weather is better for people who want to drive
2. Higher car sales volume
3. More expensive public transport cost
4. Cost of buying a car is lower
5. Higher tourism, leading to more rental of cars

## Let's look at the weather description for May

```python
# Count total traffic volume for each weather in May
get_may = df.loc[df['Month']==5]
may_weather = get_may[['weather_main', 'traffic_volume']].groupby('weather_main').sum()/sum(get_may.traffic_volume)
print(may_weather)
```

```python
# Count occurences of each weather in may
get_may['weather_main'].value_counts()/len(get_may)
```

## From the above tables we can gather the following...
1. 53% of traffic volume comes from Clear and Cloud weathers
2. 30% of weather in May are Cloud weather
3. 26% of weathers in May are Clear weathers

```python
# Sum of traffic volume by weather

plt.figure(figsize=(10,8))
plt.plot(df[['weather_main','traffic_volume']].groupby(['weather_main']).sum())
```

```python
total_traffic_volume = np.sum(df.traffic_volume)
df[['weather_main','traffic_volume']].groupby(['weather_main']).sum()/total_traffic_volume
```

### As expected, clouds and clear weathers have the highest traffic volumes. Let us see the mean traffic volume by weather

```python
# Mean of traffic volume by weather

plt.figure(figsize=(10,8))
plt.plot(df[['weather_main','traffic_volume']].groupby(['weather_main']).mean())
```

This is a surprising find! We expected to see clouds and clear weather contributing to most traffic volume because of the assumption that it is the most popular weather to drive in. However, the mean of each weather is telling us a different story. Haze has the highest mean, telling us that relatively speaking, haze may be a bigger contributor to traffic volume than clear or clouds.
